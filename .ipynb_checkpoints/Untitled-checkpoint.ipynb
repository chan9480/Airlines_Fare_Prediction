## 데이터 전처리

```python
# 데이터 불러오기
import pandas as pd
import datetime
CSV_PATH_1 = './Data_Train.csv'
CSV_PATH_2 = './Test_set.csv'
CSV_PATH_3 = './Sample_submission.csv'
target = 'Price'

# feature 수 10개.
df_train = pd.read_csv(CSV_PATH_1)
X_train = df_train.drop(target, axis=1)
y_train = df_train[target]
X_test = pd.read_csv(CSV_PATH_2)
y_test = pd.read_csv(CSV_PATH_3)

print(X_train.info())
X_train.head()
```

```python
import re
"""
전처리
1. Airline : one-hot encoding
2. Date_of_Journey : year,month feature로 변환
3. Source : one-hot encoding
4. Destination : one-hot encoding
5. Route : 삭제하자
6. Dep_Time : h*60 + min 으로 치환후 normalizing
7. Arrival_Time : 위와 동
8. Duration : 위와 동
9. Total_Stops : 숫자만 남기자 non-stop은 0
10. Additional_Info : one-hot encoding
"""

# Date_of_journey 년, 월만 남기자
def make_year(a):
    return int(a[-4:])
def make_month(a):
    return int(a[-7:-5])
X_train['year'] = X_train['Date_of_Journey'].apply(make_year)
X_train['month'] = X_train['Date_of_Journey'].apply(make_month)
X_test['year'] = X_test['Date_of_Journey'].apply(make_year)
X_test['month'] = X_test['Date_of_Journey'].apply(make_month)

print(X_train.info())
X_train.head()
```
